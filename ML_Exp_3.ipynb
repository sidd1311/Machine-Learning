```python
from google.colab import files
import pandas as pd
import numpy as np

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Extract the file name
file_name = list(uploaded.keys())[0]

# Step 3: Load the Excel file into a DataFrame
df = pd.read_excel(file_name)

# Step 4: Convert DataFrame to NumPy array (optional)
data = df.to_numpy()

# Step 5: Check the first few rows of the data
print("First few rows of the data:")
print(df.head())

```

```python
df.rename(columns={"Branch /Course": "Course"}, inplace=True)
df.rename(columns={"University Registration No. (Issued by R&S Branch)" : "Registration No."}, inplace = True)
df.rename(columns={"Date of Admission in the Present Class/Course" : "Admission Date"}, inplace = True)
df.rename(columns={"Spelling strictly as per Registration record in Capital Letters (Issued by R&S Branch)-SN" : "Student Name"}, inplace = True)
df.rename(columns={"Spelling strictly as per Registration record in Capital Letters (Issued by R&S Branch)-FN" : "Father's Name"}, inplace = True)

df.rename(columns={"Aadhaar No. of Student" : "Aadhar No."}, inplace = True)
df.rename(columns={"Updated Mobile No. of Student" : "Mobile"}, inplace = True)
df.rename(columns={"Updated E-mail ID" : "Email"}, inplace = True)
df.rename(columns={"Total Due Fee for Current Semester / (Previous semester if pending)" : "Due Fee Pending"}, inplace = True)
df.rename(columns={"Fee Paid for Current Semester / (Previous semester if pending)" :  "Fee Paid Pending"}, inplace = True)
```
