```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
data
```

```python
num_plots = sum(range(len(data.columns[:-1])))
plt.figure(figsize=(15, 12))
plot_number = 1
for i, col1 in enumerate(data.columns[:-1]):
    for col2 in data.columns[i+1:-1]:
        plt.subplot(6, 4, plot_number)
        sns.scatterplot(x=col1, y=col2, data=data)
        plt.title(f'{col1} vs {col2}')
        plot_number += 1
plt.tight_layout()
plt.show()
```

```python

sns.pairplot(data)
plt.suptitle('Pair Plot', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plot_number = 1
for col in data.columns[:-1]:
    plt.subplot(4, 4, plot_number)
    sns.histplot(data[col], kde=True, bins=15)  # Added bins parameter for better distribution clarity
    plt.title(f'{col} Distribution')
    plot_number += 1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plot_number = 1
for col in data.columns[:-1]:
    plt.subplot(4, 4, plot_number)
    sns.histplot(data[col], kde=True, bins=5)  # Added bins parameter for better distribution clarity
    plt.title(f'{col} Distribution')
    plot_number += 1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plot_number = 1
for col in data.columns[:-1]:
    plt.subplot(4, 4, plot_number)
    sns.histplot(data[col], kde=True, bins=20)  # Added bins parameter for better distribution clarity
    plt.title(f'{col} Distribution')
    plot_number += 1
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 15))
plot_number = 1
for col in data.columns[:-1]:
    plt.subplot(6, 4, plot_number)
    sns.boxplot(x='Outcome', y=col, data=data)
    plt.title(f'{col} by Outcome (Box Plot)')
    plot_number += 1
plt.tight_layout()
plt.show()
```

```python
print(f"Dataset Shape: {data.shape}")
```

```python
glucose_bp_zero = data[['Glucose', 'BloodPressure']][data['Outcome'] == 0]
print("Glucose and Blood Pressure where Outcome is 0:")
print(glucose_bp_zero)
```

```python
bmi_skin_insulin_one = data[['BMI', 'SkinThickness', 'Insulin']][data['Outcome'] == 1]
print("BMI, SkinThickness, and Insulin where Outcome is 1:")
print(bmi_skin_insulin_one)
```

```python
insulin_zero_outcome = data[['Insulin', 'Outcome']][data['Insulin'] == 0]
print("Outcome where Insulin is 0:")
print(insulin_zero_outcome)
```

```python
glucose_150_data = data[['Age', 'Outcome']][data['Glucose'] == 150]
print("Age and Outcome where Glucose is 150:")
print(glucose_150_data)
```

```python
bp_range_data = data[['Age', 'Glucose']][(data['BloodPressure'] > 70) & (data['BloodPressure'] < 90)]
print("Age and Glucose where 70 < BloodPressure < 90:")
print(bp_range_data)
```
