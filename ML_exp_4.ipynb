```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
iris = pd.read_csv("Iris.csv")

# Drop the "Id" column as it's unnecessary for visualization
iris.drop(columns=["Id"], inplace=True)

# Display the first few rows
print(iris.head())

# Pairplot - Visualizing feature relationships
sns.pairplot(iris, hue="Species", palette="husl")
plt.show()

# Boxplot - Distribution of features
plt.figure(figsize=(10,6))
sns.boxplot(data=iris, width=0.5, palette="Set2")
plt.title("Boxplot of IRIS Dataset Features")
plt.show()

# Violin Plot - Density estimation
plt.figure(figsize=(10,6))
sns.violinplot(x="Species", y="SepalLength", data=iris, palette="muted")
plt.title("Violin Plot of Sepal Length for Each Species")
plt.show()

# Scatter Plot - Sepal Length vs Petal Length
plt.figure(figsize=(8,5))
sns.scatterplot(x="SepalLength", y="PetalLength", hue="Species", data=iris, palette="coolwarm", s=100)
plt.title("Sepal Length vs Petal Length")
plt.show()
```
